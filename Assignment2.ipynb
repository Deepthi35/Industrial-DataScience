## Filtering operations

8 PANDAS FILTER METHODS TO KNOW

    * Logical operators.
    * Isin.
    * Multiple logical operators.
    * Str accessor.
    * Tilde (~).
    * Query.
    * Nlargest and nsmallest.
    * Loc and iloc.

### 1. Logical operators

```python
ds[ds.topmalebuyers > 5]
```

```python
ds[ds.buyers>400]
```

### 2. Isin

```python
names = ['France','Italie','Allemagne']
ds[ds.country.isin(names)]
```

### 3. Multiple logical operators.

And operation

```python
ds[(ds.buyers>400) & (ds.topmalebuyers > 5) ]
```

Or operation

```python
ds[(ds.buyers>400) | (ds.topmalebuyers > 5) ]
```

### 4. Str accessor

```python
ds[ds.country.str.startswith('A')]
```

```python
ds[ds.country.str.startswith('R')]
```

### 5. Tilde (~)

```python
ds[ds.country.str.contains('p')]
```

### 6. Query

```python
ds.query('buyers > 200 and femalebuyers > 5')
```

### 7. Nlargest or Nsmallest

```python
ds.nlargest(3, 'femalebuyers')
```

```python
ds.nlargest(3, 'malebuyers')
```

```python
ds.nsmallest(3, 'malebuyers')
```

### 8. Loc and Iloc

```python
ds.iloc[:,[2,3,4]].head()
```

```python
ds.loc[[2,3,4],:]
```

```python
ds.iloc[[4,3,2],[2,3,4]]
```

# write a short report on the significance of data manipulation,visulization,and analysis in extracting insights from data

Data Manipulation:
Data manipulation refers to the process of transforming raw data into a structured format that is suitable for analysis. 

    * Improved Data Quality: Data manipulation helps improve data quality by eliminating inaccuracies and inconsistencies, ensuring more reliable analysis.
    * Enhanced Data Compatibility: By integrating and transforming data from different sources, organizations can uncover relationships and patterns that would have otherwise remained hidden.
    * Efficient Analysis: Properly manipulated data sets the foundation for more efficient and accurate analysis, saving time and resources.


Data Visualization:
Data visualization is the process of presenting data in a visual format, such as charts, graphs, or maps, to facilitate understanding and gain insights. Humans are inherently visual beings, and visual representations of data enable us to identify patterns, trends, and relationships that may be difficult to discern from raw numbers.

Data Analysis:
Data analysis involves applying statistical and analytical techniques to extract meaning from data. It encompasses a range of methods, including descriptive statistics, inferential statistics, machine learning algorithms, and predictive modeling. Through data analysis, analysts can uncover correlations, identify outliers, make predictions, and uncover cause-and-effect relationships. By leveraging advanced analytics tools and techniques, organizations can make data-driven decisions, optimize processes, detect anomalies, and gain a competitive edge.

Conclusion:
Data manipulation, visualization, and analysis play a crucial role in extracting valuable insights from data. These processes enable organizations and individuals to derive meaning from raw data, make informed decisions, and gain a competitive advantage. By investing in robust data management, visualization tools, and analytical capabilities, organizations can harness the power of data and drive transformative change across various domains.
